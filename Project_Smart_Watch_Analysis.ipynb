#**3. Data Cleaning:**

* Remove the columns containing NA values.

* Remove an unwanted column "Model".

```python
#i. Remove the columns containing NA values.

new_data = smart_watch.dropna()
new_data.head(5)
```

```python
#ii. Remove "Model" column from the dataset
print(new_data.columns)
if 'Model' in new_data.columns:
   new_data.drop('Model', axis=1, inplace=True)
new_data
```

#**4. Adaptation:**

* Finding the average of 3 different columns based on the numeric values

* Then add the average column to the dataset

* Then fing the rankings based on teh averages column and print them

```python
# Select the columns
selected_columns = ['Display Size (inches)', 'Water Resistance (meters)', 'Battery Life (days)', 'Price (USD)']

# Convert the selected columns to numeric, coercing errors (non-numeric values will become NaN)
new_data[selected_columns] = new_data[selected_columns].apply(pd.to_numeric, errors='coerce')

# Calculate the average of the selected columns (ignoring NaN values)
averages = new_data[selected_columns].mean(axis=1)

# Print the result
print(averages)
```

```python
#Add the average column to the dataset

new_data['averages']=averages
new_data.head(5)
```

```python
#Find the rankings based on the averages and print them.

new_data['ranking'] = new_data['averages'].rank(ascending=False, method= 'dense')
new_data.head(5)
```

#**5. Descriptive Static Visualization:**

* Scatter Plot.

* Box Plot.

* Histogram Plot.

* Pair wise Scattered plot.

```python
# Summing Battery Life (days) by Brand
battery_by_brand = new_data.groupby('Brand')['Battery Life (days)'].sum().reset_index()

# Create the bar chart
plt.figure(figsize=(18, 10))
sns.barplot(data=battery_by_brand, x='Brand', y='Battery Life (days)', palette='viridis')

# Add title and labels
plt.title('Total Battery Life by Brand', fontsize=16)
plt.xlabel('Brand', fontsize=14)
plt.ylabel('Total Battery Life (days)', fontsize=14)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Box Plot b/w each brand of the watch and the Display Size of the watch.

plt.figure(figsize= (40,15))
sns.boxplot(data=new_data, x='Brand', y='Display Size (inches)')
plt.title('Box Plot of Display Size bt Brand')
plt.show()
```

```python
# Histogram Plot for the Resolution

plt.figure(figsize=(40,15))
plt.hist(new_data['Resolution'], bins=30, color='red', alpha=0.9)
plt.title('Histogram of Resolution')
plt.xlabel('Resolution')
plt.ylabel('Frequency')
plt.show()
```

```python
# Pair wise scattered plot

sns.set(style="ticks")
sns.pairplot(new_data, height=4, diag_kind='kde', hue='Brand', markers=['o','s','D'])
plt.show()
```
